```python
import numpy as np
baza = np.genfromtxt("mars_mchj.csv", delimiter=",",skip_header=1,dtype=None,encoding="utf-8")
baza
```

```python
fio = np.array([hodim[0] for hodim in baza])
lavozim = np.array([hodim[1] for hodim in baza])
brutto = np.array([hodim[2] for hodim in baza])
soliq = np.array([hodim[3] for hodim in baza])
jarima = np.array([hodim[4] for hodim in baza])
kpi = np.array([hodim[5] for hodim in baza])
tolov = np.array([hodim[6] for hodim in baza])
```

```python
netto = brutto* (1-soliq/100-jarima/100+kpi/100)
netto
```

```python
jami_soliq = np.sum(brutto*soliq/100)
print("Soliqlar chiqimi:",jami_soliq)
```

```python
jami_jarima = np.sum(brutto*jarima/100)
print("Jarima:",jami_jarima)
```

```python
for hodim in np.unique(lavozim):
    print("Bo'lim:", hodim,np.mean(netto[lavozim==hodim]))
```

```python
top_employee = np.argsort(netto)[-1]
print(fio[top_employee],netto[top_employee],lavozim[top_employee])
```

```python
jarima_netto = np.sum(jarima == 0)
print(jarima_netto)
```

```python
naqd = np.sum(tolov == "Naqd")/ len(tolov)*100
print(naqd)
```

```python
karta = np.sum(tolov == "Karta")/ len(tolov)*100
print(karta)
```

```python
angel = np.argmax(kpi - jarima)
print(fio[angel],netto[angel],lavozim[angel],jarima[angel])
```
